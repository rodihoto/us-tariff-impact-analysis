```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset

df = pd.read_csv("Kiel-Instituteplot_mex_can_exports_sr_ifw_en.csv")

# Prepare the plot
plt.figure(figsize=(10, 6))
bars = plt.bar(df['country_name'], df['value'], edgecolor='black')

# Add value labels on each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height,
             f"{height:.2%}", ha='center', va='bottom', fontsize=10)

# Customize plot appearance
plt.title('Impact of 25% US Tariffs (incl. oil) on Real Production (Short Run)', fontsize=14, weight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Real Production Change (%)', fontsize=12)
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.xticks(rotation=45)

# Show plot
plt.show()
```
